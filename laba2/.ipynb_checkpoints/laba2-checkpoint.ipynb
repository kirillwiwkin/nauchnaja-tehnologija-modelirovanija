# Вариант 7 
Провести анализ качества обучения НС при разных функциях
потерь:
бинарная кросс-энтропия (binary crossentropy)
категориальная кросс-энтропия (categorical crossentropy)
средний квадрат ошибок (mean squared error)

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow import keras
from tensorflow.keras.layers import Dense, Flatten
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

```python
model = keras.Sequential([
    Flatten(input_shape=(28,28,1)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
for loss in ['binary_crossentropy', 'categorical_crossentropy', 'mean_squared_error']:
    model.compile(optimizer='adam',
             loss=loss,
             metrics=['accuracy'])
    print(f'loss={loss}')
    history = model.fit(x_train, y_train_cat, batch_size=32, epochs=5, validation_split=0.2)
    model.save(f'{loss}_model')
    np.save(f'history_{loss}.npy',history.history)    
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
for loss, ax in zip(['binary_crossentropy', 'categorical_crossentropy', 'mean_squared_error'], (ax1, ax2, ax3)):
    model = keras.models.load_model(f'{loss}_model')
    history=np.load(f'history_{loss}.npy',allow_pickle='TRUE').item()
    
    ax_twinx = ax.twinx()
    plt.title(f'training with {loss}')
    plt.tight_layout()
    ax_twinx.plot(history['loss'])
    ax_twinx.plot(history['val_loss'])
    ax_twinx.legend(["loss", "val_loss"], loc ="lower right")
    (eval_loss, eval_acc) = model.evaluate(x_test, y_test_cat, verbose=False)
    print(f'evaluate with {loss.ljust(25)} - loss: {eval_loss:.4f}, accuracy: {eval_acc:.4f}')
plt.show()
```
